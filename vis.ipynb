```python
import json
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

sns.set_theme()
plt.rcParams["figure.figsize"] = (15, 7)
```

```python
def get_data(filepath = "./tmp.json") -> dict:
    data: dict = dict()
    with open(file=filepath) as f:
        data = json.load(f)
    
    return data
```

```python
data = get_data()
process_names = ["python3", "node"]
```

```python
for pname in process_names:
    cpu_dict = data.get(pname, dict()).get("cpu_precent", dict())
    cpu_dict["timestamp"] = data.get(pname, dict()).get("timestamp", list())

    cpu_df = pd.DataFrame.from_dict(cpu_dict)
    sns.lineplot(x='timestamp', y='value', hue='variable', style="variable", data=pd.melt(cpu_df, ['timestamp']))
    plt.title(f"{pname} - cpu percent")
    
    plt.show()
```

```python
for pname in process_names:
    rss_dict = data.get(pname, dict()).get("rss_usage", dict())
    rss_dict["timestamp"] = data.get(pname, dict()).get("timestamp", list())

    print(pname)

    rss_df = pd.DataFrame.from_dict(rss_dict)
    sns.lineplot(x='timestamp', y='value', hue='variable', style="variable", data=pd.melt(rss_df, ['timestamp']))
    plt.title(f"{pname} - rss usage")
    plt.show()
```
